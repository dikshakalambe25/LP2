```python
df['Roll No'] = df['Roll No'].astype(int)
#Roll No: This should be an integer. If it's in a different format, convert it to integer

df['Name'] = df['Name'].astype(str)
#Name: This should be a string. If it's not in the correct format, convert it

df[['Subject 1', 'Subject 2', 'Subject 3', 'Subject 4']] = df[['Subject 1', 'Subject 2', 'Subject 3', 'Subject 4']].apply(pd.to_numeric, errors='coerce')
#These should be numerical. Convert them to float or integer, as required

# Convert "Attendance" to numeric type, handling any errors
df['Attendance'] = pd.to_numeric(df['Attendance'], errors='coerce')
```

Data Normalization

```python
numeric_cols =['Subject 1','Subject 2',	'Subject 3',	'Subject 4',	'Attendance']
df[numeric_cols]=df[numeric_cols].fillna(method='ffill')
df

# Apply min-max normalization
df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].min()) / (df[numeric_cols].max() - df[numeric_cols].min())

# Result
print(df)
```
