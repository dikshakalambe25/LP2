A box plot shows: The median (middle value), The quartiles (spread of the data)
 and Outliers — values far from the rest are shown as dots outside the box

```python
import seaborn as sns
import matplotlib.pyplot as plt

nc=["Subject 1",	"Subject 2",	"Subject 3",	"Subject 4",	"Attendance"]
for c in nc:
  sns.boxplot(x=df[c])
  plt.title(f'box plot of {c}')
  plt.show()
```

to handle outliers after you detect them — like cleaning up the values that look "too far" from normal.
This block replaces outlier values (too small or too big) with the median of the column.



```python
for c in nc:
    Q1=df[c].quantile(0.25)
    Q3=df[c].quantile(0.75)
    IQR=Q3-Q1
    lowerbound=Q1-1.5*IQR
    upperbound=Q3+1.5*IQR
    df[c]=np.where((df[c]<lowerbound)|(df[c]>upperbound),df[c].median(),df[c])
df
```

```python
for c in nc:
    sns.boxplot(x=df[c])
    plt.title(f'Box plot of {c} after outlier treatment')
    plt.show()
```

Scan all numeric variables for outliers.
The Z-score method is a standard statistical technique used to detect outliers in numerical data.

If a Z-score is greater than a threshold (commonly 3), it suggests the value is an outlier.

You can then remove or replace these outliers, as part of data cleaning.

```python
from scipy import stats
z=np.abs(stats.zscore(df['Subject 2']))
z
```

```python
threshold=1
print(np.where(z>1))
z
```
