```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from google.colab import files
uploaded = files.upload()
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Sex', y='Age')
plt.title('Age Distribution by Gender')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Age', y='Embarked')
plt.title('Boxplot: Age vs Embarked')
plt.xlabel('Age')
plt.ylabel('Embarked')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
#observation:
"""
1) Females generally had higher survival rates compared to males, evident from the higher concentration
of data points in the "Yes" (Survived) category for women.
2) Younger passengers had a slightly higher chance of survival — this is more visible among female
children.
3) Outliers can be observed especially among males — a few very old passengers who did not survive.
4) The median age of females who survived is slightly lower compared to those who didn’t, and the IQR
(interquartile range) for females is tighter.
5) Among males, the survival age range overlaps more, and survival doesn't appear to be age-dependent as
clearly as it is for females.
"""
```
