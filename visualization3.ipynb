```python
# Create a histogram for each feature
df.hist(figsize=(10,6), edgecolor='black', grid=False)
plt.suptitle("Histograms of Iris Dataset Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#We drop the species column since it's categorical and focus on the numeric features.
#df.drop('species', axis=1) removes the species column from the dataset.
```

```python
# Create a boxplot for each feature

plt.figure(figsize=(12,8))
for i, column in enumerate(df.columns[:-1], 1): # Skip species column
    plt.subplot(2,2,i)
    sns.boxplot(y=df[column], color='darkgreen')
    plt.title(f"Boxplot of {column}")
plt.tight_layout()
plt.show()
```

```python
# Compare Distributions and identify outliers

# Observations

"""
1) Sepal Width shows a wider range and has potential outliers on both low and high ends.
2) Petal length and Petal Width show right-skewed distributions.
3) Sepal length is closer to a normal distribution.
4) Boxplots confirm outliers especially in sepal_width.
"""
```

```python
sns.histplot(x='petal_length',hue='species',data=df)

#It creates a histogram showing how the petal length is distributed.
#compare petal lengths between the species
```

```python
sns.histplot(x='petal_width',hue='species',kde=True,multiple='dodge',data=df)

#hue='species': Uses different colors to show data for each iris species.
#kde=True: Adds smooth curves (KDE lines) to show the shape of the distribution.
#multiple='dodge': Places bars side by side (not stacked), so you can easily compare species.:
#1)How petal width differs between species.
#2)Which species have wider or narrower petals.
#3)The overall spread and shape of the data for each species.
```

```python
sns.histplot(x='sepal_length',hue='species',kde=True,multiple='dodge',data=df)
```
