```python
import matplotlib.pyplot as plt

sns.histplot(data=df, x='Fare', kde=True) # Changed 'fare' to 'Fare'
plt.title('Distribution of Ticket Fares')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
#Shows how ticket fares are distributed among passengers.
#Fare distribution is right-skewed: Most passengers paid low fares, but a few paid significantly higher amounts.
#Majority paid less than ~100 units: Indicating economy or third-class tickets were more common.
#Long tail of high fares: Suggests a small number of wealthy passengers in first class
```

```python
sns.histplot(data=df, x='Fare', hue='Survived', kde=True, multiple='stack')
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.boxplot(data=df, x='Pclass', y='Fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
sns.violinplot(data=df, x='Embarked', y='Fare', hue='Survived', split=True)
plt.title('Fare Distribution by Embarked Port and Survival')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.show()
```
